```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
file_path = r'C:\Users\athar\Downloads\Zomato Data-20230302-181830287\Zomato Data\zomato.csv'
```

```python
df=pd.read_csv(file_path,encoding='ISO-8859-1')
df.head()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Aggregate rating', bins=20)
plt.xlabel('Rating')
plt.ylabel('Count')
plt.title('Distribution of Restaurant Ratings')
plt.show
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Price range', bins=5)
plt.xlabel('Price Range')
plt.ylabel('Count')
plt.title('Distribution of Restaurant Prices')
```

```python

# filter the DataFrame to only include data for Nagpur, Delhi, Mumbai, and Pune cities
cities = ['Nagpur', 'New Delhi', 'Mumbai', 'Pune']
cities_df = df[df['City'].isin(cities)]

# group the data by city and count the number of restaurants in each city
counts = cities_df.groupby('City')['Restaurant ID'].count()

# plot the restaurant counts as a bar chart
ax = counts.plot(kind='bar')
plt.xlabel('City')
plt.ylabel('Number of Restaurants')
plt.title('Number of Restaurants in Nagpur, New Delhi, Mumbai, and Pune')

# display the exact values on the bars
for i, v in enumerate(counts):
    ax.text(i - 0.1, v + 0.5, str(v), color='black')

plt.show()
```
